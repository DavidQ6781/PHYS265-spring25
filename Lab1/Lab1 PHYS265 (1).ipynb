```python
import matplotlib.pyplot as plt
import numpy as np

G = 6.67e-11 # gravitational constant in m^3/(kg s^2)

def gravitation_potential(m, xm, ym, x, y):
    '''
    This function defines the Gravitational potential energy at certain distances away from earth and returns NaN if the r = 0
    '''
    r = np.sqrt((x-xm)**2 + (y-ym)**2)
    if r == 0:
        return np.nan
    else:
        return -(G*m)/r
```

```python
# Define the absolute value of the gravitational potential as a function of x distance from earth assuming y = 0

#only plots the absolute value of the gravitation potential energy 
def abs_gravpot(X):
    return np.abs(gravitation_potential(5.9e24, 0, 0, X, 0))
x_gp = np.linspace(0, 5.7e10, 300) #creates 300 points at which to evaluate the energy at from the earth to a distance 5.7e10 m away or 1.5 x the distance from the earth to the moon
vabs = np.vectorize(abs_gravpot)  #vectorizes the function to ensure plotting
pot = vabs(x_gp) 

#1D plot of Gravitational Potential Energy
fig, ax = plt.subplots(1,1)
ax.plot(x_gp, pot)
ax.set_yscale('log')
ax.set_xlabel('Distance from Earth ($10^8 m$)')
ax.set_ylabel('Gravitational Potential (J/kg)')
ax.set_title('Gravitational Potential vs Distance from Earth')
ax.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

# Define constants
dist_earth_moon = 3.8e8           # Earth-Moon distance in m
mass_earth = 5.9 * 1e24           # Earth's mass in kg
mass_moon = 7.3 * 1e22            # Moon's mass in kg
m_commandmod = 5.5e3                 # Apollo 11 command module mass in kg

# Positions of Earth and Moon
earthx, earthy = 0, 0
moonx = dist_earth_moon / np.sqrt(2)
moony = dist_earth_moon / np.sqrt(2)

def grav_force(m1, m2, x1, y1, x2, y2):
    """
    Computes the gravitational force
    Avoids the singularity by returning NaN if the seperation between the module and Earth is less than Earth's radius.
    """
    G = 6.67e-11       # gravitational constant in m^3/(kg s^2)
    r_earth = 6.378e6  # Earth's radius in m
    
    dx = x2 - x1
    dy = y2 - y1
    r = np.sqrt(dx**2 + dy**2)
    
    if r < r_earth:
        return np.nan, np.nan
    else:
         # Gravitational force is F = G * m1 * m2 / r^2 
        # Given the vectors the formula then become F = G * m1 * m2 * (dx , dy) / r^3
        Fx = -(G * m1 * m2 * dx) / (r**3)
        Fy = -(G * m1 * m2 * dy) / (r**3)
        return Fx, Fy

def net_force(x, y):
    """
    Computes the net gravitational force on the command module at (x, y)
    due to both Earth and Moon.
    """
    Fx_e, Fy_e = grav_force(mass_earth, m_commandmod, earthx, earthy, x, y)
    Fx_m, Fy_m = grav_force(mass_moon, m_commandmod, moonx, moony, x, y)
    return Fx_e + Fx_m, Fy_e + Fy_m

# Creates a grid from -1.5*d_EM to +1.5*d_EM
grid_min = -1.5 * dist_earth_moon
grid_max =  1.5 * dist_earth_moon
x_vals3 = np.linspace(grid_min, grid_max, 500)
y_vals3 = np.linspace(grid_min, grid_max, 500)
X, Y = np.meshgrid(x_vals3, y_vals3)

# Vectorize the net_force function so that it can operate on arrays
v_net_force = np.vectorize(net_force)
Fx_field, Fy_field = v_net_force(X, Y)

# Compute force magnitude for color
F_mag = np.hypot(Fx_field, Fy_field)

# vmin and vmax for LogNorm (must be positive)
F_mag_valid = F_mag[F_mag > 0]
if F_mag_valid.size == 0:
    vmin_val = 1e-15
    vmax_val = 1.0
else:
    vmin_val = np.nanmin(F_mag_valid)
    vmax_val = np.nanmax(F_mag_valid)

# Plot the streamplot
fig, ax4 = plt.subplots(1,1)
stream = ax4.streamplot(X, Y, Fx_field, Fy_field, color=F_mag, cmap='plasma', norm=colors.LogNorm(vmin=vmin_val, vmax=vmax_val),density=2, linewidth=1)

ax4.set_xlabel('X (m)')
ax4.set_ylabel('Y (m)')
ax4.set_title('Gravitational Force on Apollo Command Module\n(Earth-Moon System)')
ax4.set_xlim(grid_min, grid_max)
ax4.set_ylim(grid_min, grid_max)

# Marks for Eart and Moon
ax4.scatter(earthx, earthy, color='blue', marker='o', s=100, label="Earth")
ax4.scatter(moonx, moony, color='gray', marker='o', s=70, label="Moon")

# colorbar
cbar = fig.colorbar(stream.lines, ax=ax4)
cbar.set_label('Force Magnitude (N)')
plt.legend()
ax4.set_aspect('equal')
plt.tight_layout()
plt.show()
```

```python
from scipy.integrate import quad

# Define Variables

m0 = 2.8e6 #kg initial(wet) mass
mf = 7.5e5 #kg final(dry) mass
m = 1.3e4 #kg/s flow rate
ve = 2.4e3 #m/s exhaust velocity
g = 9.81 #m/s^2 

#Equations

T = (m0 - mf) / m  #Rocket burn time

def velocity_rocket(t, g, m0, mt, ve):
    """
    Using equation returns the velocity of the rocket at time t
    """
    return ve * np.log(m0/(m0 - m*t)) - (g*t)

altitude_h, e = quad(lambda x:velocity_rocket(x,g,m0,m,ve), 0 ,T) #Integral from t=0 to t=T to find total altitude

v_burnout = velocity_rocket(T, g, m0, m, ve) #Time at which Stage 1 reaches burnout
t = np.linspace(0, T, 500) #Time data points
v = velocity_rocket(t, g, m0, m, ve) #Speed of the rocket through out time t = 0 and time t = T

#Plotting to examine how velocity looks over time
fig, ax5 = plt.subplots(1,1)
ax5.plot(t,v, color='r', label = 'Velocity')
ax5.set_ylabel('Velocity (m/s)')
ax5.set_xlabel('Time (s)')
ax5.set_title('Saturn V velocity over time until burnout')
plt.legend()
plt.tight_layout()
ax5.grid()

print(f' Altitude at burnout = {altitude_h:.2f} meters')
```
